```python
import csv
import pandas as pd

# Suponha que você tenha três DataFrames chamados dataframe2003, dataframe2013 e dataframe2023
# Substitua os nomes dos DataFrames pelos seus nomes reais


#Renomeando coluna do datasets
dataframe2023.rename(columns={'Hora UTC':'HORA (UTC)', 'RADIACAO GLOBAL (Kj/mý)': 'RADIACAO GLOBAL (KJ/mý)'}, inplace=True)
dataframe2003.rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (MM)': 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'},inplace=True)
#print(dataframe2023)
#print(dataframe2023.columns.values)

#Convertendo coluna DATA para datetime
#dataframe2003['DATA'] = pd.to_datetime(dataframe2003['DATA'], format='%d/%m/%Y', errors='coerce').dt.strftime('%Y/%m/%d')
#print(dataframe2003)
#dataframe2013['DATA'] = pd.to_datetime(dataframe2013['DATA'], format='%d/%m/%Y', errors='coerce').dt.strftime('%Y/%m/%d')
#dataframe2023['DATA'] = pd.to_datetime(dataframe2023['DATA'], format='%d/%m/%Y', errors='coerce').dt.strftime('%Y/%m/%d')

dataframe2003['DATA'] = pd.to_datetime(dataframe2003['DATA'], format='mixed')
#print(dataframe2003)
#dataframe2003.drop('DATA', axis=1, inplace=True)

dataframe2013['DATA'] = pd.to_datetime(dataframe2013['DATA'],format='mixed')
#print(dataframe2013)
#dataframe2013.drop('DATA', axis=1, inplace=True)

dataframe2023['DATA'] = pd.to_datetime(dataframe2023['DATA'],format='mixed')

#print(dataframe2023)
#dataframe2023.drop('DATA', axis=1, inplace=True)
#dataframe2023['HORA (UTC)'] = dataframe2023['HORA (UTC)'].fillna('0000 UTC')

#dataframe2023['HORA (UTC)'] = pd.to_datetime(dataframe2023['HORA (UTC)'],format='ISO8601')
dataframe2023['HORA (UTC)'] = pd.to_datetime(dataframe2023['HORA (UTC)'], format='%H%M UTC', errors='coerce')
dataframe2023['HORA (UTC)'] = dataframe2023['HORA (UTC)'].dt.strftime('%H:%M')
#print(dataframe2023['HORA (UTC)'].value_counts())

#dataframe2023.columns.tolist()
#print(type(df_concat.columns))
#print((dataframe2023.columns))
#print(dataframe2023.info())
#print(dataframe2023['HORA (UTC)'])
#dataframe2023['HORA (UTC)'] = pd.to_datetime(dataframe2023['HORA (UTC)'],format='mixed', utc=True)#.strftime('%H:%M'))
#dataframe2023['HORA (UTC)'] = dataframe2023['HORA (UTC)'].date.dt.strftime('%H:%M')

# Leia os arquivos CSV ou substitua pelo método adequado para carregar seus DataFrames
#dataframe2003 = pd.read_csv('./datasets/INMET_CO_MT_A901_CUIABA_01-01-2003_A_31-12-2003.CSV', sep=';', encoding='ISO-8859-1')
#dataframe2013 = pd.read_csv('./datasets/INMET_CO_MT_A901_CUIABA_01-01-2013_A_31-12-2013.CSV', sep=';', encoding='ISO-8859-1')
#dataframe2023 = pd.read_csv('./datasets/INMET_CO_MT_A901_CUIABA_01-01-2023_A_31-12-2023.CSV', sep=';', encoding='ISO-8859-1')
#print(dataframe2023)

# Concatene os DataFrames ao longo do eixo das linhas
df_concat = pd.concat([dataframe2003, dataframe2013, dataframe2023])
#print(df_concat)

df_concat['DATA'] = pd.to_datetime(df_concat['DATA'])
df_concat['Ano'] = df_concat['DATA'].dt.year
df_concat['Mes'] = df_concat['DATA'].dt.month
df_concat['Dia'] = df_concat['DATA'].dt.day

multiIndex = pd.MultiIndex.from_frame(df_concat[['Ano', 'Mes', 'Dia']])#transforma essas colunas em multi index

df_concat.set_index(multiIndex, inplace=True)
#df_concat.drop('DATA', axis=1, inplace=True)

df_concat.drop(['Ano','Mes','Dia'], axis=1, inplace=True) # substitui a instrução abaixo (iloc)
#df_concat.drop('Mês', axis=1, inplace=True) # mesmo de cima
#df_concat.drop('Dia', axis=1, inplace=True) # mesmo de cima

#df_concat = df_concat.iloc[:, :-3] # slice do dataframe sem 3 ultimas colunas (ano, mes e dia)

#df_concat.to_csv('saida3.csv', index=False)
# Visualize o DataFrame com MultiIndex no índice
#print(df_concat)

#df_concat.columns.tolist()
#print(type(df_concat.columns))


arrayCols=[]
for cols in df_concat.columns:
    tupla=(cols,"")
    if(cols.find("PRESS")!=-1):
        tupla=("PRESSAO",cols)
    elif(cols.find("TEMPERATURA")!=-1):
        tupla=("TEMPERATURA",cols)
    elif(cols.find("UMIDADE")!=-1):
        tupla=("UMIDADE",cols)
    elif(cols.find("VENTO")!=-1):
        tupla=("VENTO",cols)
    arrayCols.append(tupla)

#arrayCols

df_concat.columns = pd.MultiIndex.from_tuples(arrayCols)

#for cols in df_concat.columns:
    #df_concat[cols] = df_concat[cols].replace('-9999.0', pd.nan)
df_concat.replace({-9999.0: None}, inplace=True)

#print(df_concat)
df_concat.to_csv('saida3.csv', index=False,na_rep='NaN')
#print(dataframe2023.head())
print((df_concat.columns.tolist()))


```
